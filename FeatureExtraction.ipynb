```python
from glob import glob
fakedata_dir = '/content/drive/MyDrive/AudioDataset/training/fake/fake'
fakedata_files = glob(fakedata_dir+'**/*.wav')
realdata_dir = '/content/drive/MyDrive/AudioDataset/training/real/real'
realdata_files = glob(realdata_dir+'**/*.wav')
print(f'Number of fake audios : {len(fakedata_files)}')
print(f'Number of real audios : {len(realdata_files)}')
```

```python
import librosa
from librosa import feature
import numpy as np
fn_list_i = [
 feature.mfcc,
 feature.melspectrogram,
 feature.spectral_centroid,
 feature.spectral_bandwidth,
 feature.spectral_rolloff
]
 
fn_list_ii = [
 feature.rms,
 feature.zero_crossing_rate
]
def get_feature_vector(y,sr): 
   feat_vect_i = [ np.mean(funct(y,sr)) for funct in fn_list_i]
   feat_vect_ii = [ np.mean(funct(y)) for funct in fn_list_ii] 
   feature_vector = feat_vect_i + feat_vect_ii 
   return feature_vector
```

```python
fakedata_feature=[]
for file in fakedata_files:
  y, sr = librosa.load(file, sr=None)
  feature_vector=get_feature_vector(y,sr)
  fakedata_feature.append(feature_vector)
  print('.', end= " ")
```

```python
print(fakedata_feature)
```

```python
realdata_feature=[]
for file in realdata_files:
  y, sr = librosa.load(file, sr=None)
  feature_vector=get_feature_vector(y,sr)
  realdata_feature.append(feature_vector)
  print('.', end= " ")
```

```python
print(realdata_feature)
```

```python
import csv

fake_output = 'fake_00.csv'
real_output = 'real_00.csv'

header =[
    'mfcc',
    'melspectogram',
    'spectral_centroid',
    'spectral_bandwidth',
    'spectral_rolloff',
    'rms',
    'zero_crossing_rate'
]

with open(fake_output,'+w') as f:
  csv_writer = csv.writer(f, delimiter = ',')
  csv_writer.writerow(header)
  csv_writer.writerows(fakedata_feature)

with open(real_output,'+w') as f:
  csv_writer = csv.writer(f, delimiter = ',')
  csv_writer.writerow(header)
  csv_writer.writerows(realdata_feature)
```
